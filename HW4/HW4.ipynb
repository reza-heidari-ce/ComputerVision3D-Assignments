```python
def estimateF(x1,x2):
    # Return the fundamental matrix F (3 by 3), based on two sets of homogeneous 2D points x1 and x2.
    # Input: x1,x2 numpy ndarray (3 by N) containing matching 2D homogeneous points.
    # Output: F numpy ndarray (3 by 3) containing the fundamental matrix.
    N = x1.shape[1]
    O = np.zeros(shape=(N, 9))
    for i in range(N):
        x, y, z = x1[:, i]
        x_prime, y_prime, z_prime = x2[:, i]
        O[i] = np.array([x * x_prime, x * y_prime, x, y * x_prime, y * y_prime, y, x_prime, y_prime, 1])
    U, S, V = np.linalg.svd(O)
    e_min = V[-1].reshape(3, 3)
    U, S, V = np.linalg.svd(e_min)
    S[-1] = 0
    F = U @ np.diag(S) @ V
    return F

def estimateFnorm(x1,x2):
    # Return the fundamental matrix F (3 by 3), based on two sets of homogeneous 2D points x1 and x2.
    # Input: x1,x2 numpy ndarray (3 by N) containing matching 2D homogeneous points.
    # Output: F numpy ndarray (3 by 3) containing the fundamental matrix based on normalized homogeneous points.
    T = np.array([[2 / np.mean(x1[0]), 0, 0], [0, 2 / np.mean(x1[1]), 0], [0, 0, 1]]) @ np.array([[1, 0, -np.mean(x1[0])], [0, 1, -np.mean(x1[1])], [0, 0, 1]]) 
    T_prime = np.array([[2 / np.mean(x2[0]), 0, 0], [0, 2 / np.mean(x2[1]), 0], [0, 0, 1]]) @ np.array([[1, 0, -np.mean(x2[0])], [0, 1, -np.mean(x2[1])], [0, 0, 1]]) 
    x1_norm = T @ x1
    x2_norm = T_prime @ x2
    F = estimateF(x1_norm, x2_norm)
    F = T_prime.T @ F @ T
    return F
 
def vgg_F_from_P(P1,P2):
    # Return the fundamental matrix F (3 by 3), based on two camera parameter arrays.
    # Input: P1, P2 numpy ndarray (3 by 4) containing intrinsic and extrinsic parameters.
    # Output: F numpy ndarray (3 by 3) containing the fundamental matrix.
    X=[]
    Y=[]
    X.append(P1[[1,2],:])
    X.append(P1[[2,0],:])
    X.append(P1[[0,1],:])
    Y.append(P2[[1,2],:])
    Y.append(P2[[2,0],:])
    Y.append(P2[[0,1],:])
    F=np.zeros([3,3])
    
    for i in range(3):
        for j in range(3):
            M=np.concatenate([X[j],Y[i]])
            F[i,j]=np.linalg.det(M)
    return F
```
